```
"""
CIRCLES
"""

import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75,75,50,-50])

plt.axis('on')
plt.grid(True)

plt.arrow(0,0,20,0,head_length=4,head_width=3,color='k')
plt.arrow(0,0,0,20,head_length=4,head_width=3,color='k')

plt.text(16,-3,'x')
plt.text(-5,17,'y')

#------------------------------------green circle
xc=0
yc=0
r=40

p1=0*np.pi/180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc+r*np.cos(p1)
ylast = yc+r*np.sin(p1)

for p in np.arange(p1, p2 + dp, dp):
    x = xc+r*np.cos(p)
    y = yc+r*np.sin(p)

    if p > 90 * np.pi / 180 and p < 270 * np.pi / 180:
        plt.plot([xlast, x], [ylast, y], color='g', linestyle=':')
    else:
        plt.plot([xlast, x], [ylast, y], color='g')

    xlast = x
    ylast = y

plt.scatter(xc, yc, s=15, color='g')

# ________________________________________red circle
xc = -20
yc = -20
r = 10

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)

for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    plt.plot([xlast, x], [ylast, y], linewidth=4, color='r')
    xlast = x
    ylast = y

plt.scatter(xc, yc, s=15, color='r')

# ________________________________________purple circle
xc = 20
yc = 20
r = 50

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)

for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    plt.plot([xlast, x], [ylast, y], linewidth=2, color=(.8, 0, .8))
    xlast = x
    ylast = y

plt.scatter(xc, yc, color=(.5, 0, .5))

# ________________________________________blue disc
xc = -53
yc = -30
r1 = 0
r2 = 10
dr = 1

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r1 * np.cos(p1)
ylast = yc + r1 * np.sin(p1)

for r in np.arange(r1, r2, dr):
    for p in np.arange(p1, p2 + dp, dp):
        x = xc + r * np.cos(p)
        y = yc + r * np.sin(p)
        plt.plot([xlast, x], [ylast, y], linewidth=2, color=(0, 0, .8))
        xlast = x
        ylast = y

plt.show()
```

```
# Listing 2-7. program ELLIPSES
"""
ELLIPSES
"""

import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75, 75, 50, -50])
plt.axis('on')
plt.grid(True)

plt.arrow(0, 0, 60, 0, head_length=4, head_width=3, color='k')
plt.arrow(0, 0, 0, -45, head_length=4, head_width=3, color='k')  # Arah panah y terbalik

plt.text(58, -3, 'x')
plt.text(-5, -44, 'y')  # Posisi teks y disesuaikan

#---------------------------------------------------red ellipse
a = 40
b = 20
p = np.linspace(0, 2 * np.pi, 100)  # Membuat array sudut yang lebih halus
x_red = a * np.cos(p)
y_red = b * np.sin(p)
plt.plot(x_red, y_red, color='r')

#--------------------------------------green ellipse
a = 20
b = 40
p = np.linspace(0, 2 * np.pi, 100)  # Membuat array sudut yang lebih halus
x_green = a * np.cos(p)
y_green = b * np.sin(p)
plt.plot(x_green, y_green, color='g')

#---------------------------blue ellipse (filled)
a = 5
b = 15
p = np.linspace(0, 2 * np.pi, 100)  # Membuat array sudut yang lebih halus
x_blue = a * np.cos(p)
y_blue = b * np.sin(p)
plt.fill(x_blue, y_blue, color='navy')  # Menggunakan fill untuk membuat ellipse berwarna solid

plt.show()
```

```
# Listing 2-9
"""
2DTRANSLATION
"""

import numpy as np
import matplotlib.pyplot as plt

x1 = -10
x2 = 140
y1 = 0  # Perubahan pada batas y
y2 = 90  # Perubahan pada batas y
plt.axis([x1, x2, y2, y1])  # Urutan y terbalik agar sesuai gambar

plt.axis('on')
plt.grid(True)

plt.title('Translation')

#----------------------------------- Triangle
x_triangle = [20, 30, 40, 20]
y_triangle = [40, 20, 40, 40]
plt.plot(x_triangle, y_triangle, color='k')

#---------------------------------------translate triangle dx=60
x_triangle_translated_dx = [val + 60 for val in x_triangle]
plt.plot(x_triangle_translated_dx, y_triangle, color='g')

#---------------------------------------translate triangle dy=40
x_triangle_translated_dy = x_triangle
y_triangle_translated_dy = [val + 40 for val in y_triangle]
plt.plot(x_triangle_translated_dy, y_triangle_translated_dy, color='r')

#--------------------------------------box
x_box = [0, 0, 5, 5, 0]
y_box = [55, 50, 50, 55, 55]
plt.plot(x_box, y_box, 'b')

#-------------------------------------translate box
y_box_translated = [55, 50, 50, 55, 55]
for x_offset in np.arange(0, 130, 10):
    x_box_translated = [val + x_offset for val in x_box]
    plt.plot(x_box_translated, y_box_translated, 'b')

plt.show()
```

```
# Listing 2-11. Program 2DROTRECTANGLE
"""
2DROTRECTANGLE
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([-10, 150, 100, -10])
plt.axis('on')
plt.grid(True)

# ------------------------------------axes
plt.arrow(0, 0, 40, 0, head_length=4, head_width=2, color='b')
plt.arrow(0, 0, 0, 40, head_length=4, head_width=2, color='b')
plt.text(30, -3, 'Xg', color='b')
plt.text(-8, 34, 'Yg', color='b')

xc = 75  # ----------------------------center of rotation
yc = 50
plt.plot([xc - 40, xc + 60], [yc, yc], linewidth=2, color='grey')  # --X  # Ketebalan garis diubah menjadi 2
plt.plot([xc, xc], [yc - 40, yc + 45], linewidth=2, color='grey')  # --Y  # Ketebalan garis diubah menjadi 2
plt.text(127, 48, 'X')
plt.text(70, 90, 'Y')

plt.scatter(xc, yc, s=20, color='k')  # -plot center of rotation
plt.text(70, 49, 'c')

# --------------------------------------------------------define function rotz
def rotz(xp, yp, rz):
    c11 = np.cos(rz)
    c12 = -np.sin(rz)
    c21 = np.sin(rz)
    c22 = np.cos(rz)
    xpp = xp * c11 + yp * c12  # --relative to xc,yc
    ypp = xp * c21 + yp * c22
    xg = xc + xpp  # -- relative to xg,yg
    yg = yc + ypp
    return [xg, yg]

# ---------------------------------------------------------plot unrotated rectangle
# -------------rectangle corner coordinates in X,Y system
xp1 = -20
xp2 = +20
xp3 = +20
xp4 = -20
yp1 = -5
yp2 = -5
yp3 = +5
yp4 = +5

plt.text(50, 45, '1')  # -----------label
plt.text(97, 45, '2')
plt.text(97, 57, '3')
plt.text(50, 57, '4')
plt.scatter(xp1 + xc, yp1 + yc, s=10, color='k')
plt.scatter(xp2 + xc, yp2 + yc, s=10, color='k')
plt.scatter(xp3 + xc, yp3 + yc, s=10, color='k')
plt.scatter(xp4 + xc, yp4 + yc, s=10, color='k')

# ---------------------------------------------------------plot unrotated rectangle
xg1 = xc + xp1  # -------------corner coordinates in Xg,Yg system
yg1 = yc + yp1
xg2 = xc + xp2
yg2 = yc + yp2
xg3 = xc + xp3
yg3 = yc + yp3
xg4 = xc + xp4
yg4 = yc + yp4

xg = [xg1, xg2, xg3, xg4, xg1]
yg = [yg1, yg2, yg3, yg4, yg1]

plt.plot((xg), (yg), color='k')

# ----------------------------------------rotate rectangle corner coordinates
rz = 45
rz = rz * np.pi / 180

# ------------------------------------------------point 1
xp = xp1
yp = yp1
[xg, yg] = rotz(xp, yp, rz)
[xg1, yg1] = [xg, yg]

# ------------------------------------------------point 2
xp = xp2
yp = yp2
[xg, yg] = rotz(xp, yp, rz)
[xg2, yg2] = [xg, yg]

# ------------------------------------------------point 3
xp = xp3
yp = yp3
[xg, yg] = rotz(xp, yp, rz)
[xg3, yg3] = [xg, yg]

# ---------------------------------------------------point 4
xp = xp4
yp = yp4
[xg, yg] = rotz(xp, yp, rz)
[xg4, yg4] = [xg, yg]

# ----------------------------------------------------plot rotated rectangle
xg = [xg1, xg2, xg3, xg4, xg1]
yg = [yg1, yg2, yg3, yg4, yg1]

plt.plot(xg, yg, color='r')

plt.show()
```
